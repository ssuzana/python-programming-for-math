# Parametric Surfaces

If we want to plot a surface, then we need to construct 2d arrays containing the locations of the x and y coordinates, and a 2d array containing the height of the surface. For structured data (ie, where the x and y coordinates lie on a regular grid) the meshgrid function helps.


$$x = (1-u)(3+\cos v) \cos 4\pi u$$
$$y = (1-u)(3+\cos v) \sin 4\pi u$$
$$z=3u+(1-u)\sin v$$

```python
# Define the parametric equations
u = np.linspace(0, 1, 100)
v = np.linspace(0, 2 * np.pi, 100)
u, v = np.meshgrid(u, v)

x = (1 - u) * (3 + np.cos(v)) * np.cos(4 * np.pi * u)
y = (1 - u) * (3 + np.cos(v)) * np.sin(4 * np.pi * u)
z = 3 * u + (1 - u) * np.sin(v)

# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='plasma')

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis', labelpad=-0.5)


# Show plot
plt.title('Parametric Surface Plot')
plt.show()
```

# Quadric Surfaces 

## Elliptic Paraboloid 

$$\frac{z}{c} = \frac{x^2}{a^2} +\frac{y^2}{b^2}$$

```python
# Define the parameters
a = 2  # Semi-axis in x-direction
b = 3  # Semi-axis in y-direction
c = 1  # Scaling factor for z

# Generate grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z values based on the equation of the elliptic paraboloid
z = (x**2 / a**2 + y**2 / b**2) * c

# Plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Show plot
plt.title('Elliptic Paraboloid: ' + r'$\frac{z}{c} = \frac{x^2}{a^2} + \frac{y^2}{b^2}$')
plt.show()
```

## Hyperbolic Paraboloid

$$\frac{z}{c} = \frac{x^2}{a^2} - \frac{y^2}{b^2}$$

```python
# Define the parameters
a = 2  # Semi-axis in x-direction
b = 3  # Semi-axis in y-direction
c = 1  # Scaling factor for z

# Generate grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z values based on the equation of the hyperbolic paraboloid
z = (x**2 / a**2) - (y**2 / b**2) * c

# Plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels and title with LaTeX
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.title('Hyperbolic Paraboloid: ' + r'$\frac{z}{c} = \frac{x^2}{a^2} - \frac{y^2}{b^2}$')

# Show plot
plt.show()
```

## Cone

$$\frac{z^2}{c} = \frac{x^2}{a^2} + \frac{y^2}{b^2}$$


This code will generate a 3D plot of a quadric surface cone using the specified parameters for $a, b,$ and $c$. It calculates $z$ values by taking the square root of the equation of the cone and then plots the upper and lower halves of the cone.

```python
# Define the parameters
a = 1 # Semi-axis in x-direction
b = 1  # Semi-axis in y-direction
c = 1  # Scaling factor for z

# Generate grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z values based on the equation of the quadric surface cone
z = np.sqrt((x**2 / a**2) + (y**2 / b**2)) * np.sqrt(c)

# Plot the surface
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
#ax.plot_surface(x, y, -z, cmap='viridis')  # Plot the other half of the cone

# Set labels and title with LaTeX
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')


plt.title('Cone: ' + r'$\frac{z^2}{c} = \frac{x^2}{a^2} + \frac{y^2}{b^2}$')

# Show plot
plt.show()
```

```python
# Define cone parameters
a = 1
b = 1
c = 1

# Create data for the cone
theta = np.linspace(0, 2*np.pi, 100)  
r = np.linspace(-5, 5, 100) 
Theta, R = np.meshgrid(theta, r) 
X = a * R * np.cos(Theta)
Y = b * R * np.sin(Theta)
Z = c * R


# Create a 3D surface plot for the cone
fig_cone = plt.figure(figsize=(6,8))
ax_cone = fig_cone.add_subplot(projection='3d')
ax_cone.plot_surface(X, Y, Z, cmap='plasma')

plt.title('Cone: ' + r'$\frac{z^2}{c} = \frac{x^2}{a^2} + \frac{y^2}{b^2}$') 

plt.show()
```

## Ellipsoid

The ellipsoid may be parameterized in several ways, which are simpler to express when the ellipsoid axes coincide with coordinate axes. A common choice is



$${\displaystyle {\begin{aligned}x&=a\sin(\theta )\cos(\varphi ),\\y&=b\sin(\theta )\sin(\varphi ),\\z&=c\cos(\theta ),\end{aligned}}\,\!}$$
where

$${\displaystyle 0\leq \theta \leq \pi ,\qquad 0\leq \varphi <2\pi .}$$
These parameters may be interpreted as spherical coordinates, where  is the polar angle and  is the azimuth angle of the point $(x, y, z)$ of the ellipsoid.

```python
# Define ellipsoid parameters
a = 5
b = 1
c = 2

# Create a mesh grid
theta = np.linspace(0, np.pi, 50)    # Polar angle
phi = np.linspace(0, 2 * np.pi, 100)   # Azimuthal angle
Theta, Phi = np.meshgrid(theta, phi)

# Calculate the coordinates of points on the ellipsoid surface
X = a * np.sin(Theta) * np.cos(Phi)
Y = b * np.sin(Theta) * np.sin(Phi)
Z = c * np.cos(Theta)

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='plasma')

ax.set_zlim(-3,3)

plt.title("Ellipsoid")
plt.show()
```

## Hyperbolic Paraboloid

$$x = acosh(u)cos(v)$$
$$y = bcosh(u)sin(v)$$
$$z = csinh(u)$$

```python
# Define the parametric equations
def parametric_eq(u, v, a, b, c):
    x = a * np.cosh(u) * np.cos(v)
    y = b * np.cosh(u) * np.sin(v)
    z = c * np.sinh(u)
    return x, y, z

# Define range of parameters u and v
u = np.linspace(-3, 3, 100)
v = np.linspace(0, 2 * np.pi, 100)
u, v = np.meshgrid(u, v)

# Define parameters a, b, and c
a = 2
b = 3
c = 1

# Compute the parametric values
x, y, z = parametric_eq(u, v, a, b, c)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Labeling axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Hyperbolic Paraboloid')

plt.show()
```

## Hyperboloid of Two Sheets

$$x = a \sinh(u) \cos(v)$$
$$y = b \sinh(u) \sin(v)$$
$$z = c \cosh(u)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the parametric equations for both sheets
def parametric_eq(u, v, a, b, c):
    x = a * np.sinh(u) * np.cos(v)
    y = b * np.sinh(u) * np.sin(v)
    z = np.empty_like(u)
    z[u >= 0] = c * np.cosh(u[u >= 0])  # Positive sheet
    z[u < 0] = -c * np.cosh(-u[u < 0])  # Negative sheet
    return x, y, z

# Define range of parameters u and v
u = np.linspace(-3, 3, 100)
v = np.linspace(0, 2 * np.pi, 100)
u, v = np.meshgrid(u, v)

# Define parameters a, b, and c
a = 2
b = 3
c = 1

# Compute the parametric values
x, y, z = parametric_eq(u, v, a, b, c)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Labeling axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Parametric Plot of Hyperboloid of Two Sheets')

plt.show()
```
